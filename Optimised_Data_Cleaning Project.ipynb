```python
def clean_data(file_path):
    # Reading the CSV file
    data = pd.read_csv(file_path)
    
    # Deleting the first 5 rows
    data = data.iloc[5:]
    
    # Removing 'Unnamed: 4' column
    data = data.drop('Unnamed: 4', axis=1)
    
    # Renaming all the columns
    data = data.rename({
        'Practice Report For Data Cleaning': 'Program Code',
        'Unnamed: 1': 'Program Name',
        'Unnamed: 2': 'Level',
        'Unnamed: 3': 'Number of Students',
        'Unnamed: 5': 'School',
        'Unnamed: 6': 'Deposit Amount'
    }, axis=1)
    
    # Removing rows and columns with all null values
    data = data.dropna(how='all').dropna(how='all', axis=1)
    
    # Removing rows where 'Program Name' is 'TOTAL HIGH DEMAND'
    data = data[data['Program Name'] != 'TOTAL HIGH DEMAND']
    
    # Forward fill to replicate values in the rows below them
    data = data.fillna(method='ffill')
    
    # Removing the first 6 rows
    data = data.iloc[6:]
    
    # Resetting the index
    data = data.reset_index(drop=True)
    
    # Changing the data type for 'Deposit Amount'
    data['Deposit Amount'] = data['Deposit Amount'].replace('[\$,]', '', regex=True).astype(int)
    
    # Changing the data type for 'Number of Students'
    data['Number of Students'] = data['Number of Students'].astype(int)
    
    return data
```

```python
clean_data = clean_data('Raw file for Test for A3.csv')
clean_data
```
